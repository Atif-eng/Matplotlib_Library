# Fill between plot

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color='r')
plt.fill_between(x=[2,4],y1=2,y2=4)

plt.show()
```

```python
plt.plot(x,y,color='r')
plt.fill_between(x,y)

plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color='r')
plt.fill_between(x,y,where=(x>=2) & (x<=4),alpha=0.5)

plt.show()
```
