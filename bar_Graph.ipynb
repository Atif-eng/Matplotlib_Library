# Bar Graph in Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[6,3,4,5]
c=["g",'y',"b",'r']
plt.bar(x,y,color=c,width=0.5,edgecolor='m',linewidth=2,linestyle=":",alpha=0.5)
plt.show()
```

```python
from matplotlib import pyplot as plt

x=["Python","C","Java","HTML","C#"]
y=[90,70,85,40,100]
z=[25,45,21,10,65]

plt.bar(x,y,width=0.7,color="r",edgecolor="r",linewidth=2,linestyle=":",alpha=0.3,label="Atif_1")
plt.bar(x,z,width=0.7,color="y",edgecolor="r",linewidth=2,linestyle=":",alpha=0.3,label="Atif_2")
# plt.bar(x,y,width=0.5,color="black",align='edge')
# plt.bar(x,y,width=0.3,color="blue",align='center')

plt.xlabel("Languages",fontsize=20)
plt.ylabel("Percentage",fontsize=20)
plt.title("matplotlib.ipynb",fontsize=15)

plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x=["Python","C","Java","HTML","C#"]
y=[90,70,85,40,100]
z=[25,45,21,10,65]

width=0.3

p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Languages",fontsize=20)
plt.ylabel("Percentage",fontsize=20)
plt.title("matplotlib.ipynb",fontsize=15)

plt.bar(x,y,width,color="r",edgecolor="r",linewidth=2,linestyle=":",alpha=0.5,label="Atif_1")
plt.bar(p1,z,width,color="y",edgecolor="r",linewidth=2,linestyle=":",alpha=0.5,label="Atif_2")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt

x=["Python","C","Java","HTML","C#"]
y=[90,70,85,40,100]
z=[25,45,21,10,65]

width=0.5
p=[j+width for j in range(len(x))]

plt.ylabel("Languages",fontsize=20)
plt.xlabel("Percentage",fontsize=20)
plt.title("matplotlib.ipynb",fontsize=15)
c=["r","y","b","g","m"]

plt.barh(x,y,width,color="r",label="Graph_XY",alpha=0.5)
plt.barh(p,z,width,color="b",label="Graph_XZ",alpha=0.5)

plt.legend()
plt.show()
```
