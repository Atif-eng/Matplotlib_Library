# Box Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
x=[10,20,30,40,50,60,70]

plt.title('Box Plot',fontsize=25)
plt.boxplot(x,widths=0.2,labels=['Python'],patch_artist=True,showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,patch_artist=True,sym='g+',boxprops={'color':'r'},capprops={'color':'g'},whiskerprops={'color':'m'},flierprops={'markeredgecolor':'r'})
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]

plt.title('Box Plot',fontsize=20)
plt.boxplot(x,widths=0.2,labels=['Python'],whis=2)
plt.show()
```

```python
plt.boxplot(x,vert=False)
plt.show()
```

```python
plt.boxplot(x,notch=True)
plt.show()
```
