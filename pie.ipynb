# Pie Graph

```python
from matplotlib import pyplot as plt
x=[10,20,30,40,50]
y=['C','Java','HTML','C#','Python']
ex=[0.0,0.0,0.0,0.0,0.2]
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',shadow=True)
plt.show()
```

```python
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',shadow=True,radius=1,labeldistance=1.2,startangle=90)
plt.show()
```

```python
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',shadow=True,textprops={"fontsize":15})
plt.show()
```

```python
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',shadow=True,counterclock=False)
plt.show()
```

```python
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',shadow=True,wedgeprops={"linewidth":5,'edgecolor':'Pink'},rotatelabels=True)
plt.show()
```

```python
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',shadow=True,)
plt.title('PIE Graph',fontsize=20)
plt.legend(loc=3)
plt.show()
```

`Dot PIE chart`

```python
plt.pie([1])
plt.show()
```

`Multiple PIE chart`

```python
from matplotlib import pyplot as plt
x=[10,20,30,40,50]
x1=[50,40,30,20,10]
y=['C','Java','HTML','C#','Python']

plt.pie(x,labels=y,radius=1)
plt.pie(x1,radius=0.5)

plt.show()
```

`Ring PIE chart`

```python
from matplotlib import pyplot as plt
x=[10,20,30,40,50]
x1=[50,40,30,20,10]
y=['C','Java','HTML','C#','Python']

plt.pie(x,labels=y,radius=1)
plt.pie([1],radius=0.7,colors="white")

plt.show()
```
