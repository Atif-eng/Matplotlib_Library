# Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
y=[0,1,2,3,4,5,6]
x=[2,3,1,4,6,5,3]
colors=[12,45,27,10,67,80,54]

plt.scatter(x,y,s=400,c=colors,marker="o",cmap='BrBG')#s=size

t=plt.colorbar()
t.set_label('colorbar',fontsize=20)
plt.title('Scatter Graph',fontsize=15)

plt.show()
```
