`Hist Plot Using DataSet`

```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("C:\\Users\\HP\\OneDrive\\Documentos\\python\\jupyter_note_book\\12.csv")

# Plotting histograms for each column in the DataFrame
for col in data.keys():
    plt.figure(figsize=(8, 6))  
    plt.hist(data[col],color='r' ,bins=20, edgecolor='black', alpha=0.6)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)#Grid lines for better readability
    plt.show()
```
